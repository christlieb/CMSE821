# 2D Poisson Equation on the Unit Square — Separation of Variables (Detailed)

**Problem (homogeneous Dirichlet).**

We seek \(u=u(x,y)\) such that
$$
\begin{aligned}
-\,\Delta u &\equiv -\big(u_{xx}+u_{yy}\big)=f(x,y)
&& \text{in } (0,1)\times(0,1),\\
u &= 0 && \text{on } \partial\big((0,1)\times(0,1)\big).
\end{aligned}
$$

---

## 1) Separation Ansatz and Reduction

Assume a product form
$$
u(x,y)=X(x)\,Y(y).
$$

Substitute into \( -\Delta u = f \). The separation method first solves the **homogeneous eigenproblem**
for the Laplacian (the inhomogeneous right–hand side will be handled by an expansion):
$$
-\,\Delta \phi = \lambda\,\phi, \qquad \phi|_{\partial\Omega}=0.
$$
With \(\phi(x,y)=X(x)Y(y)\), we get
$$
-\frac{X''(x)}{X(x)} - \frac{Y''(y)}{Y(y)}=\lambda
\quad\Longrightarrow\quad
-\frac{X''}{X}=\mu,\ \ -\frac{Y''}{Y}=\nu,\ \ \mu+\nu=\lambda.
$$

---

## 2) Spatial Eigenproblems with Dirichlet BCs

The boundary condition \(u=0\) on \(x=0,1\) and \(y=0,1\) implies
\(X(0)=X(1)=0\) and \(Y(0)=Y(1)=0\). The 1D eigenproblems are
$$
X''+\mu X=0,\quad X(0)=X(1)=0,
\qquad
Y''+\nu Y=0,\quad Y(0)=Y(1)=0.
$$

These are classical Sturm–Liouville problems with the nontrivial solutions
$$
X_m(x)=\sin(m\pi x),\ \ \mu_m=(m\pi)^2;\qquad
Y_n(y)=\sin(n\pi y),\ \ \nu_n=(n\pi)^2;\qquad m,n=1,2,\dots
$$

Hence the **2D eigenpairs** are
$$
\phi_{mn}(x,y)=\sin(m\pi x)\sin(n\pi y),
\qquad
\lambda_{mn}=\mu_m+\nu_n=(m^2+n^2)\pi^2.
$$

The family \(\{\phi_{mn}\}\) is orthogonal in \(L^2((0,1)^2)\) and automatically satisfies the Dirichlet BCs.

---

## 3) Orthogonality and Normalization

For \(m,p\ge 1\) and \(n,q\ge 1\),
$$
\int_0^1\!\!\int_0^1 \phi_{mn}\,\phi_{pq}\,dx\,dy
= \left(\int_0^1 \sin(m\pi x)\sin(p\pi x)\,dx\right)
  \left(\int_0^1 \sin(n\pi y)\sin(q\pi y)\,dy\right)
= \frac{1}{4}\,\delta_{mp}\delta_{nq}.
$$

It is convenient to work with these **unnormalized** basis functions and keep the factor \(1/4\) in mind.

---

## 4) Expansions for \(f\) and \(u\), and Coefficient Equations

Expand both \(f\) and \(u\) in the sine basis:
$$
\begin{aligned}
f(x,y) &= \sum_{m=1}^{\infty}\sum_{n=1}^{\infty} F_{mn}\,\phi_{mn}(x,y),\\
u(x,y) &= \sum_{m=1}^{\infty}\sum_{n=1}^{\infty} U_{mn}\,\phi_{mn}(x,y).
\end{aligned}
$$

Using \(-\Delta \phi_{mn}=\lambda_{mn}\phi_{mn}\) and orthogonality, substituting into
\(-\Delta u=f\) yields (mode by mode)
$$
\lambda_{mn}\,U_{mn}=F_{mn}\quad\Longrightarrow\quad
U_{mn}=\frac{F_{mn}}{\lambda_{mn}}.
$$

The (unnormalized) **sine coefficients** of \(f\) are
$$
F_{mn}=4\int_{0}^{1}\!\!\int_{0}^{1}
f(x,y)\,\sin(m\pi x)\,\sin(n\pi y)\,dy\,dx.
$$

---

## 5) Final Series Solution (Dirichlet on the Unit Square)

Putting everything together,
$$
\boxed{
\ u(x,y)=\sum_{m=1}^{\infty}\sum_{n=1}^{\infty}
\frac{F_{mn}}{\lambda_{mn}}\,
\sin(m\pi x)\sin(n\pi y),
\qquad
\lambda_{mn}=(m^2+n^2)\pi^2,\quad
F_{mn}=4\!\int_{0}^{1}\!\!\int_{0}^{1} f\,\phi_{mn}\,dy\,dx.
}
$$

This series converges rapidly for smooth \(f\); the boundary condition \(u=0\) holds termwise.

---

## 6) Worked Example: Piecewise-Constant Forcing on a Sub-Rectangle

Let
$$
f(x,y)=\begin{cases}
1, & x\in\big[\tfrac{1}{4},\tfrac{1}{2}\big],\quad y\in\big[\tfrac{1}{2},\tfrac{3}{4}\big],\\[2pt]
0, & \text{otherwise}.
\end{cases}
$$

Compute the sine coefficients by separating the integrals:
$$
\begin{aligned}
F_{mn}
&=4\int_{1/4}^{1/2}\!\sin(m\pi x)\,dx
     \int_{1/2}^{3/4}\!\sin(n\pi y)\,dy\\[4pt]
&=\frac{4}{mn\,\pi^2}\,
\Big[\cos\!\big(\tfrac{m\pi}{4}\big)-\cos\!\big(\tfrac{m\pi}{2}\big)\Big]\,
\Big[\cos\!\big(\tfrac{n\pi}{2}\big)-\cos\!\big(\tfrac{3n\pi}{4}\big)\Big],
\end{aligned}
$$
using \(\displaystyle \int_a^b \sin(k\pi s)\,ds=\frac{\cos(k\pi a)-\cos(k\pi b)}{k\pi}\).

Therefore,
$$
U_{mn}=\frac{F_{mn}}{\lambda_{mn}}
=\frac{4}{mn\,(m^2+n^2)\,\pi^4}\,
\Big[\cos\!\big(\tfrac{m\pi}{4}\big)-\cos\!\big(\tfrac{m\pi}{2}\big)\Big]\,
\Big[\cos\!\big(\tfrac{n\pi}{2}\big)-\cos\!\big(\tfrac{3n\pi}{4}\big)\Big],
$$
and the solution is the sine series
$$
\boxed{
\ u(x,y)=\sum_{m=1}^{\infty}\sum_{n=1}^{\infty}
\frac{4}{mn\,(m^2+n^2)\,\pi^4}\,
\Big[\cos\!\big(\tfrac{m\pi}{4}\big)-\cos\!\big(\tfrac{m\pi}{2}\big)\Big]\,
\Big[\cos\!\big(\tfrac{n\pi}{2}\big)-\cos\!\big(\tfrac{3n\pi}{4}\big)\Big]\,
\sin(m\pi x)\sin(n\pi y).
}
$$

**Notes.**

- The series satisfies \(u=0\) on \(\partial\Omega\) termwise.  
- Modes with \((m,n)\) that make either bracket vanish simply do not contribute.  
- Coefficients decay like \(O((m^2+n^2)^{-3/2})\) for this box \(f\) (two jumps), yielding a rapidly convergent series.

---

```python
"""
2D Poisson: -Δ u = f on (0,1)^2, u|_{∂Ω}=0
RHS: f = 1 on [1/4,1/2] × [1/2,3/4], 0 otherwise.
Solved on an n×n interior grid using 2D DST-I.
"""

import numpy as np
import matplotlib.pyplot as plt

# Try SciPy FFT interface first
try:
    from scipy.fft import dst, idst
    _SCIPY_OK = True
except Exception:
    try:
        from scipy.fftpack import dst, idst
        _SCIPY_OK = True
    except Exception:
        _SCIPY_OK = False

def poisson_dirichlet_unit_square_mask(n=256):
    """
    Solve -Δ u = f with f=1 on [1/4,1/2]×[1/2,3/4], else 0.
    Returns interior grids x,y and solution U (n×n).
    """
    if not _SCIPY_OK:
        raise ImportError("Requires SciPy's dst/idst. Install with `pip install scipy`.")

    h = 1.0 / (n + 1)                 # interior spacing
    x = np.linspace(h, 1.0 - h, n)    # interior x
    y = np.linspace(h, 1.0 - h, n)    # interior y
    X, Y = np.meshgrid(x, y, indexing="ij")

    # RHS: indicator of the rectangle [1/4,1/2] × [1/2,3/4]
    F = ((X >= 0.25) & (X <= 0.50) & (Y >= 0.50) & (Y <= 0.75)).astype(float)

    # Forward 2D DST-I (orthonormal)
    F_hat = dst(dst(F, type=1, axis=0, norm='ortho'), type=1, axis=1, norm='ortho')

    # Discrete Laplacian eigenvalues for DST-I basis
    jj = np.arange(1, n + 1)
    s = np.sin(np.pi * jj / (2.0 * (n + 1)))
    lam = (4.0 / h**2) * (s[:, None]**2 + s[None, :]**2)

    U_hat = F_hat / lam

    # Inverse 2D DST-I
    U = idst(idst(U_hat, type=1, axis=0, norm='ortho'), type=1, axis=1, norm='ortho')
    return x, y, U, F

if __name__ == "__main__":
    x, y, U, F = poisson_dirichlet_unit_square_mask(n=256)

    # Plot solution surface
    X, Y = np.meshgrid(x, y, indexing="ij")
    from mpl_toolkits.mplot3d import Axes3D  # noqa: F401
    fig = plt.figure()
    ax = fig.add_subplot(111, projection="3d")
    ax.plot_surface(X, Y, U, linewidth=0, antialiased=True)
    ax.set_title("Solution u(x,y) for f=1 on [1/4,1/2]×[1/2,3/4], 0 o.w.")
    ax.set_xlabel("x"); ax.set_ylabel("y")
    plt.tight_layout()
    plt.show()

    # Optional: visualize RHS mask
    plt.figure()
    plt.imshow(F.T, origin="lower", extent=(x[0], x[-1], y[0], y[-1]), aspect="auto")
    plt.title("RHS mask f(x,y)")
    plt.xlabel("x"); plt.ylabel("y")
    plt.tight_layout()
    plt.show()

    # Print a simple norm of u
    print("||u||_inf =", float(np.max(np.abs(U))))
```

# 2D Heat Equation with Homogeneous Neumann BCs: Separation of Variables (Detailed)

**Problem.**

$$
\begin{aligned}
u_t &= \kappa\,(u_{xx}+u_{yy}) && \text{in } (0,1)\times(0,1)\times(0,\infty),\\
u_x(0,y,t)&=u_x(1,y,t)=0, \quad u_y(x,0,t)=u_y(x,1,t)=0,\\
u(x,y,0)&=f(x,y).
\end{aligned}
$$

---

## 1) Separation Ansatz and Reduction

Assume a product form
$$
u(x,y,t)=X(x)\,Y(y)\,T(t).
$$

Substitute into the PDE and divide by \(X Y T\):
$$
\frac{T'(t)}{\kappa\,T(t)}
= \frac{X''(x)}{X(x)} + \frac{Y''(y)}{Y(y)}
= -\lambda.
$$

Hence there exists a separation constant \(\lambda\ge 0\) and
$$
\frac{X''}{X} = -\mu,\qquad
\frac{Y''}{Y} = -\nu,\qquad
\mu+\nu=\lambda.
$$

---

## 2) Spatial Eigenproblems with Neumann BCs

From the boundary conditions \(\partial_x u=0\) at \(x=0,1\) and \(\partial_y u=0\) at \(y=0,1\), we get
$$
X'(0)=X'(1)=0,\qquad Y'(0)=Y'(1)=0.
$$

Thus \(X\) and \(Y\) solve the 1D Sturm–Liouville problems
$$
X''+\mu X=0,\quad X'(0)=X'(1)=0;\qquad
Y''+\nu Y=0,\quad Y'(0)=Y'(1)=0.
$$

**Solve for \(X\).** The general solution is \(X(x)=A\cos(\sqrt{\mu}\,x)+B\sin(\sqrt{\mu}\,x)\).
The Neumann condition at \(x=0\) gives \(X'(0)=B\sqrt{\mu}=0\), hence \(B=0\).
The condition at \(x=1\) requires \(X'(1)=-A\sqrt{\mu}\sin(\sqrt{\mu})=0\).
For nontrivial \(A\), we need \(\sin(\sqrt{\mu})=0 \Rightarrow \sqrt{\mu}=m\pi\), \(m=0,1,2,\dots\).
Therefore
$$
X_m(x)=\cos(m\pi x),\qquad \mu_m=(m\pi)^2,\qquad m=0,1,2,\dots
$$
(The \(m=0\) mode \(X_0\equiv 1\) is allowed under Neumann BCs.)

**Solve for \(Y\).** Analogously,
$$
Y_n(y)=\cos(n\pi y),\qquad \nu_n=(n\pi)^2,\qquad n=0,1,2,\dots
$$

**2D eigenpairs.**
$$
\phi_{mn}(x,y)=\cos(m\pi x)\cos(n\pi y),\qquad
\lambda_{mn}=(m^2+n^2)\pi^2.
$$

---

## 3) Temporal Factors

With \(u=XYT\) and \(\lambda=\lambda_{mn}\),
$$
T'_{mn}(t) + \kappa\,\lambda_{mn}\,T_{mn}(t)=0
\quad\Rightarrow\quad
T_{mn}(t)=
\begin{cases}
e^{-\kappa\,\lambda_{mn}\,t}, & (m,n)\neq(0,0),\\
1, & (m,n)=(0,0)\ \text{(mean mode)}.
\end{cases}
$$

The \((0,0)\) mode corresponds to \(\lambda_{00}=0\) and carries the conserved spatial mean.

---

## 4) Orthogonality and Coefficients

The cosine basis is orthogonal on \([0,1]\):
$$
\int_0^1 \cos(m\pi x)\cos(p\pi x)\,dx =
\begin{cases}
1, & m=p=0,\\[2pt]
\frac{1}{2}, & m=p\ge 1,\\[2pt]
0, & m\ne p.
\end{cases}
$$

Hence an \(L^2\) expansion (cosine–cosine series) takes the form
$$
\begin{aligned}
u(x,y,t)
&=\sum_{m=0}^\infty\sum_{n=0}^\infty
a_{mn}\,e^{-\kappa\,\lambda_{mn} t}\,\phi_{mn}(x,y),\\
a_{mn}
&=\alpha_m\,\alpha_n \int_0^1\!\!\int_0^1
f(x,y)\,\cos(m\pi x)\cos(n\pi y)\,dy\,dx,\\
\alpha_0&=1,\qquad \alpha_k=2\ (k\ge 1).
\end{aligned}
$$

---

## 5) Worked Example: Box Initial Data

Let
$$
f(x,y)=\begin{cases}
1, & x\in\big[\tfrac14,\tfrac34\big],\ \ y\in\big[\tfrac14,\tfrac34\big],\\[2pt]
0, & \text{otherwise}.
\end{cases}
$$

Separate the coefficient integral:
$$
a_{mn}=\alpha_m\alpha_n\,I_m\,J_n,
\quad
I_m:=\int_{1/4}^{3/4}\!\cos(m\pi x)\,dx,\quad
J_n:=\int_{1/4}^{3/4}\!\cos(n\pi y)\,dy.
$$

Compute the 1D pieces:
$$
\begin{aligned}
I_m&=
\begin{cases}
\frac{1}{2}, & m=0,\\[6pt]
\dfrac{\sin\!\big(\tfrac{3m\pi}{4}\big)-\sin\!\big(\tfrac{m\pi}{4}\big)}{m\pi}, & m\ge 1,
\end{cases}
&
J_n&=
\begin{cases}
\frac{1}{2}, & n=0,\\[6pt]
\dfrac{\sin\!\big(\tfrac{3n\pi}{4}\big)-\sin\!\big(\tfrac{n\pi}{4}\big)}{n\pi}, & n\ge 1.
\end{cases}
\end{aligned}
$$

The solution is therefore
$$
\begin{aligned}
u(x,y,t)
&=\sum_{m=0}^\infty\sum_{n=0}^\infty
\alpha_m\alpha_n\,I_m\,J_n\,
e^{-\kappa\,\pi^2(m^2+n^2)\,t}\,
\cos(m\pi x)\cos(n\pi y),\\
&\text{with the mean mode } a_{00}=I_0J_0=\tfrac{1}{4}\ \text{(time-independent)}.
\end{aligned}
$$

**Notes.**  
1) Neumann BCs admit the constant eigenfunction; the spatial average \(\int u\) is conserved.  
2) Each nonconstant mode decays like \(e^{-\kappa\pi^2(m^2+n^2)t}\), so high frequencies smooth out fastest.  
3) The series converges to \(f\) in \(L^2\) as \(t\downarrow 0\) and is smooth for any \(t>0\).


```python
# 2D Heat Equation on (0,1)x(0,1) with Homogeneous Neumann BCs
# Separation-of-variables solution for the box initial condition:
#   f(x,y) = 1 on [1/4, 3/4] x [1/4, 3/4], and 0 otherwise.
#
# u(x,y,t) = Σ_{m=0..M} Σ_{n=0..N} a_{mn} e^{-κ π^2 (m^2+n^2) t} cos(mπx) cos(nπy)
# where a_{mn} = α_m α_n I_m J_n, with α_0=1, α_k=2 (k>=1),
# I_m = ∫_{1/4}^{3/4} cos(mπx) dx, J_n = ∫_{1/4}^{3/4} cos(nπy) dy.

import numpy as np
import matplotlib.pyplot as plt

# -------- Parameters --------
kappa = 0.1          # thermal diffusivity κ
M = 30               # number of cosine modes in x (includes m=0)
N = 30               # number of cosine modes in y (includes n=0)
Nx = 201             # grid points in x for visualization
Ny = 201             # grid points in y for visualization
times = [0.0, 0.002, 0.01, 0.05]  # times to visualize
# ----------------------------

def alpha(k: int) -> int:
    """Normalization factor for unnormalized cosine basis on [0,1]:
       ∫_0^1 cos(kπx)^2 dx = 1 (k=0) or 1/2 (k>=1).
       Coefficient formula uses α_0=1, α_k=2 (k>=1)."""
    return 1 if k == 0 else 2

def box_integral_I(m: int) -> float:
    """I_m = ∫_{1/4}^{3/4} cos(mπ x) dx (closed form)."""
    if m == 0:
        return 0.5
    return (np.sin(3*m*np.pi/4) - np.sin(m*np.pi/4)) / (m*np.pi)

def box_integral_J(n: int) -> float:
    """J_n = ∫_{1/4}^{3/4} cos(nπ y) dy (closed form)."""
    if n == 0:
        return 0.5
    return (np.sin(3*n*np.pi/4) - np.sin(n*np.pi/4)) / (n*np.pi)

def build_coefficients(M: int, N: int) -> np.ndarray:
    """Build the coefficient matrix a[m,n] for the box initial data under Neumann BCs."""
    Im = np.array([box_integral_I(m) for m in range(M+1)], dtype=float)          # (M+1,)
    Jn = np.array([box_integral_J(n) for n in range(N+1)], dtype=float)          # (N+1,)
    am = np.array([alpha(m) for m in range(M+1)], dtype=float)                   # (M+1,)
    an = np.array([alpha(n) for n in range(N+1)], dtype=float)                   # (N+1,)
    # a[m,n] = α_m α_n I_m J_n  (outer products)
    A = (am * Im)[:, None] * (an * Jn)[None, :]                                  # (M+1, N+1)
    return A

def eval_solution_on_grid(A: np.ndarray, kappa: float, t: float, Nx: int, Ny: int):
    """Evaluate u(x,y,t) on an (Nx x Ny) grid in [0,1]^2 given coefficients A ((M+1)x(N+1))."""
    M = A.shape[0] - 1
    N = A.shape[1] - 1
    x = np.linspace(0.0, 1.0, Nx)                                                # (Nx,)
    y = np.linspace(0.0, 1.0, Ny)                                                # (Ny,)

    # Mode indices
    m_col = np.arange(M+1)[:, None]                                              # (M+1,1)
    n_col = np.arange(N+1)[:, None]                                              # (N+1,1)
    n_row = np.arange(N+1)[None, :]                                              # (1,N+1)

    # Cosine basis values on grids
    Cx = np.cos(m_col * np.pi * x[None, :])                                      # (M+1, Nx)
    Cy = np.cos(n_col * np.pi * y[None, :])                                      # (N+1, Ny)

    # Exponential decay factors for each mode (m,n)
    lambdas = (m_col**2 + n_row**2) * (np.pi**2)                                  # (M+1, N+1)
    decay = np.exp(-kappa * lambdas * t)                                         # (M+1, N+1)

    # Assemble u(x,y,t) = Cx^T * (A ⊙ decay) * Cy
    Adecay = A * decay                                                           # (M+1, N+1)
    U = Cx.T @ (Adecay @ Cy)                                                     # (Nx, Ny)

    return x, y, U

def plot_heatmap(x, y, U, title: str = "") -> None:
    """Plot a heatmap of U(x,y)."""
    plt.figure()
    extent = (x[0], x[-1], y[0], y[-1])
    plt.imshow(U.T, origin="lower", extent=extent, aspect="equal")
    plt.colorbar(label="u(x,y)")
    plt.xlabel("x"); plt.ylabel("y")
    plt.title(title)
    plt.tight_layout()
    plt.show()

# ---- Build coefficients once (depends only on initial condition) ----
A = build_coefficients(M, N)

# Report mean (should be 1/4 for the box). The (0,0) coefficient equals the mean.
mean_from_coeffs = float(A[0, 0])
print(f"Mean of initial condition from a_00 = {mean_from_coeffs:.6f} (expected 0.25)")

# ---- Visualize solution at requested times ----
for t in times:
    x, y, U = eval_solution_on_grid(A, kappa, t, Nx, Ny)
    plot_heatmap(x, y, U, title=f"2D Heat Eq (Neumann) — t = {t}")

# (Optional) quick numerical check: the spatial mean should stay ≈ a_00 for all t
for t in times:
    _, _, Uc = eval_solution_on_grid(A, kappa, t, 121, 121)
    mean_grid = float(Uc.mean())
    print(f"Approx. mean(u(·,·,{t})) ≈ {mean_grid:.6f}")
```

# 2D Wave Equation on the Unit Square — Separation of Variables (Detailed)

**Problem (homogeneous Dirichlet).**

We seek \(u=u(x,y,t)\) such that
$$
\begin{aligned}
u_{tt} &= c^2\,(u_{xx}+u_{yy}) && \text{in } (0,1)\times(0,1)\times(0,\infty),\\
u &= 0 && \text{on } \partial\big((0,1)\times(0,1)\big)\ \text{for all } t\ge 0,\\
u(x,y,0) &= f(x,y),\qquad u_t(x,y,0)=g(x,y).
\end{aligned}
$$

---

## 1) Separation Ansatz and Reduction

Assume a product form
$$
u(x,y,t)=X(x)\,Y(y)\,T(t).
$$

Substitute into the PDE and divide by \(X\,Y\,T\):
$$
\frac{T''(t)}{c^2\,T(t)}
= \frac{X''(x)}{X(x)} + \frac{Y''(y)}{Y(y)}
= -\lambda.
$$

Thus there exists a separation constant \(\lambda\ge 0\) and
$$
\frac{X''}{X}=-\mu, \qquad \frac{Y''}{Y}=-\nu, \qquad \mu+\nu=\lambda.
$$

---

## 2) Spatial Eigenproblems with Dirichlet BCs

The boundary condition \(u=0\) on \(x=0,1\) and \(y=0,1\) implies
\(X(0)=X(1)=0\) and \(Y(0)=Y(1)=0\). The 1D problems are
$$
X''+\mu X=0,\quad X(0)=X(1)=0;\qquad
Y''+\nu Y=0,\quad Y(0)=Y(1)=0.
$$

Nontrivial solutions:
$$
X_m(x)=\sin(m\pi x),\ \ \mu_m=(m\pi)^2;\qquad
Y_n(y)=\sin(n\pi y),\ \ \nu_n=(n\pi)^2;\qquad m,n=1,2,\dots
$$

Hence the **2D eigenpairs** are
$$
\phi_{mn}(x,y)=\sin(m\pi x)\sin(n\pi y),\qquad
\lambda_{mn}=(m^2+n^2)\pi^2.
$$

Orthogonality (unnormalized basis):
$$
\int_0^1\!\!\int_0^1 \phi_{mn}\,\phi_{pq}\,dx\,dy
= \tfrac{1}{4}\,\delta_{mp}\delta_{nq}.
$$

---

## 3) Temporal Factors

For each \((m,n)\), \(T_{mn}\) solves
$$
T_{mn}''(t) + c^2\lambda_{mn}\,T_{mn}(t)=0,
\qquad \omega_{mn}:=c\sqrt{\lambda_{mn}}=c\pi\sqrt{m^2+n^2}.
$$
Thus
$$
T_{mn}(t)=A_{mn}\cos(\omega_{mn} t)+B_{mn}\sin(\omega_{mn} t).
$$

---

## 4) General Series Solution and Coefficients

Expand in the eigenbasis:
$$
u(x,y,t)=\sum_{m=1}^{\infty}\sum_{n=1}^{\infty}
\big(a_{mn}\cos(\omega_{mn} t)+b_{mn}\sin(\omega_{mn} t)\big)\,\phi_{mn}(x,y).
$$

Use the initial data and orthogonality to determine coefficients:
- From \(u(x,y,0)=f(x,y)\),
  $$
  a_{mn}=4\int_0^1\!\!\int_0^1 f(x,y)\,\sin(m\pi x)\sin(n\pi y)\,dy\,dx.
  $$
- From \(u_t(x,y,0)=g(x,y)\),
  $$
  b_{mn}=\frac{4}{\omega_{mn}}\int_0^1\!\!\int_0^1 g(x,y)\,\sin(m\pi x)\sin(n\pi y)\,dy\,dx.
  $$

These formulas already incorporate the \(\tfrac{1}{4}\) coming from the unnormalized
orthogonality of \(\phi_{mn}\).

---

## 5) Worked Example: Box Initial Displacement, Zero Initial Velocity

Let
$$
f(x,y)=\begin{cases}
1, & x\in\big[\tfrac14,\tfrac34\big],\ \ y\in\big[\tfrac14,\tfrac34\big],\\[2pt]
0, & \text{otherwise},
\end{cases}
\qquad
g(x,y)\equiv 0.
$$

Compute \(a_{mn}\) by separating variables:
$$
\begin{aligned}
a_{mn}
&=4\!\left(\int_{1/4}^{3/4}\sin(m\pi x)\,dx\right)
     \left(\int_{1/4}^{3/4}\sin(n\pi y)\,dy\right)\\[4pt]
&=\frac{4}{mn\,\pi^2}\,
\Big[\cos\!\big(\tfrac{m\pi}{4}\big)-\cos\!\big(\tfrac{3m\pi}{4}\big)\Big]\,
\Big[\cos\!\big(\tfrac{n\pi}{4}\big)-\cos\!\big(\tfrac{3n\pi}{4}\big)\Big],
\end{aligned}
$$
using \(\displaystyle \int_a^b \sin(k\pi s)\,ds=\frac{\cos(k\pi a)-\cos(k\pi b)}{k\pi}\).
Since \(g\equiv 0\), we have \(b_{mn}=0\).

Therefore the solution is
$$
\boxed{
\ u(x,y,t)=\sum_{m=1}^{\infty}\sum_{n=1}^{\infty}
\frac{4}{mn\,\pi^2}\,
\Big[\cos\!\big(\tfrac{m\pi}{4}\big)-\cos\!\big(\tfrac{3m\pi}{4}\big)\Big]\,
\Big[\cos\!\big(\tfrac{n\pi}{4}\big)-\cos\!\big(\tfrac{3n\pi}{4}\big)\Big]\,
\cos\!\big(c\pi\sqrt{m^2+n^2}\,t\big)\,
\sin(m\pi x)\sin(n\pi y).
}
$$

**Remarks.**
- The boundary condition \(u=0\) is enforced termwise by the sine basis.  
- Energy is conserved mode-by-mode (no damping), while the spectrum determines wave speeds.  
- Only modes with nonvanishing brackets contribute; symmetry of the box data often forces many \(a_{mn}\) to zero.


```python
# 2D Wave Equation on (0,1)x(0,1) with Homogeneous Dirichlet BCs
# Separation-of-variables solution for the box initial displacement:
#   f(x,y) = 1 on [1/4, 3/4] x [1/4, 3/4], and 0 otherwise,
#   g(x,y) = 0 (initial velocity).

import numpy as np
import matplotlib.pyplot as plt

# -------- Parameters you can tweak --------
c = 1.0                # wave speed
M = 50                 # number of sine modes in x (m = 1..M)
N = 50                 # number of sine modes in y (n = 1..N)
Nx = 201               # grid points in x for visualization
Ny = 201               # grid points in y for visualization
times = [0.0, 0.1, 0.3, 0.6]  # times to visualize
# ------------------------------------------

def build_coefficients(M: int, N: int) -> np.ndarray:
    """
    Build the coefficient matrix a[m-1, n-1] = a_{mn} for the box initial displacement
    f = 1 on [1/4,3/4]^2, 0 elsewhere, under Dirichlet BCs.
    """
    m = np.arange(1, M+1, dtype=float)  # (M,)
    n = np.arange(1, N+1, dtype=float)  # (N,)

    # ∫_{1/4}^{3/4} sin(kπ s) ds = [cos(kπ/4) - cos(3kπ/4)] / (kπ)
    Im = (np.cos(m*np.pi/4.0) - np.cos(3.0*m*np.pi/4.0)) / (m*np.pi)  # (M,)
    Jn = (np.cos(n*np.pi/4.0) - np.cos(3.0*n*np.pi/4.0)) / (n*np.pi)  # (N,)

    # a_{mn} = 4 * Im * Jn  (outer product)
    A = 4.0 * np.outer(Im, Jn)  # (M, N)
    return A

def eval_solution_on_grid(A: np.ndarray, c: float, t: float, Nx: int, Ny: int):
    """
    Evaluate u(x,y,t) on an (Nx x Ny) grid in [0,1]^2 given coefficients A of shape (M,N).
    Implements:
      u(x,y,t) = Sx^T @ ( (A * cos(ω t)) @ Sy )
    where Sx[m,x] = sin(mπx), Sy[n,y] = sin(nπy), ω_{mn} = cπ sqrt(m^2+n^2).
    """
    M, N = A.shape
    x = np.linspace(0.0, 1.0, Nx)  # (Nx,)
    y = np.linspace(0.0, 1.0, Ny)  # (Ny,)

    # Mode indices (column for basis eval, row for frequency grid)
    m_col = np.arange(1, M+1, dtype=float)[:, None]   # (M,1)
    n_col = np.arange(1, N+1, dtype=float)[:, None]   # (N,1)  <-- use this for Sy
    n_row = np.arange(1, N+1, dtype=float)[None, :]   # (1,N)  <-- use this for omega

    # Basis values
    Sx = np.sin(m_col * np.pi * x[None, :])           # (M, Nx)
    Sy = np.sin(n_col * np.pi * y[None, :])           # (N, Ny)  <-- FIXED broadcasting

    # Frequencies and time factor
    omega = c * np.pi * np.sqrt(m_col**2 + n_row**2)  # (M, N)
    timefac = np.cos(omega * t)                       # (M, N)

    # Assemble u(x,y,t)
    A_t = A * timefac                                 # (M, N)
    U = Sx.T @ (A_t @ Sy)                             # (Nx, Ny)

    return x, y, U

def plot_heatmap(x: np.ndarray, y: np.ndarray, U: np.ndarray, title: str = "") -> None:
    """Plot a heatmap of U(x,y)."""
    plt.figure()
    extent = (x[0], x[-1], y[0], y[-1])
    plt.imshow(U.T, origin="lower", extent=extent, aspect="equal")
    plt.colorbar(label="u(x,y,t)")
    plt.xlabel("x"); plt.ylabel("y")
    plt.title(title)
    plt.tight_layout()
    plt.show()

# ---- Build coefficients (depends only on initial displacement) ----
A = build_coefficients(M, N)

# Preview of t=0 (expect Gibbs oscillations)
x0, y0, U0 = eval_solution_on_grid(A, c, t=0.0, Nx=Nx, Ny=Ny)
plot_heatmap(x0, y0, U0, title="Wave Eq (Dirichlet) — t = 0 (truncated series)")

# ---- Visualize solution at requested times ----
for t in times:
    x, y, U = eval_solution_on_grid(A, c, t, Nx, Ny)
    plot_heatmap(x, y, U, title=f"Wave Eq (Dirichlet) — t = {t}")

# Simple boundary sanity check (Dirichlet -> ~0 on edges with truncation error)
print("Edge sample values (|u| should be small):")
print(f"left  edge max |u| ≈ {np.abs(U[:, 0]).max():.3e}")
print(f"right edge max |u| ≈ {np.abs(U[:, -1]).max():.3e}")
print(f"bottom edge max |u| ≈ {np.abs(U[0, :]).max():.3e}")
print(f"top    edge max |u| ≈ {np.abs(U[-1, :]).max():.3e}")
```
